# Capturing first data set 

```python
import json
import pandas as pd
tweets_data_path = 'tweets_SK.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
        
print(len(tweets_data))

tweets = pd.DataFrame()
tweets['text'] = [tweet.get('text','') for tweet in tweets_data]
tweets['lang'] = [tweet.get('lang','') for tweet in tweets_data]
tweets_by_lang = tweets['lang'].value_counts()
```

```python
tweets
```

Cleaning the first data set by removing words like "RT", "@" and tweets made in other languages 

```python
import re
tweets['text'] = tweets['text'].str.replace('@', '', case=False)
tweets['text'] = tweets['text'].str.replace('RT', '', case=False)
tweets['text'] = tweets['text'].loc[tweets['lang']=='en']
tweets
```

Cleaning first data set by removing other symbols and urls

```python
tweets['text'] = tweets['text'].str.replace('#', '', case=False)
tweets['text'] = tweets['text'].str.replace(':', '', case=False)
tweets['text'] = tweets['text'].str.replace(',', '', case=False)
tweets.text.str.encode('utf-8').str.decode('ascii', 'ignore')
tweets['text'] = tweets['text'].str.replace('http\S+|www.\S+', '', case=False)
tweets
```

# Capturing second data set

```python
import json
import pandas as pd
tweets_data_path = 'tweets_SK_2.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
        
print(len(tweets_data))

tweets2 = pd.DataFrame()
tweets2['text'] = [tweet.get('text','') for tweet in tweets_data]
tweets2['lang'] = [tweet.get('lang','') for tweet in tweets_data]
tweets_by_lang = tweets2['lang'].value_counts()
```

Cleaning second data set as the first data set

```python
tweets2 = pd.DataFrame()
tweets2['text'] = tweets['text'].loc[tweets['lang']=='en']
tweets2['text'] = tweets['text'].str.replace('@', '', case=False)
tweets2['text'] = tweets['text'].str.replace('RT', '', case=False)
import re
def clean(text):
    regex = r'RT'
    return re.sub(regex, "", text)
tweets2['text'] = list(tweets2['text'].apply(lambda tweet: clean(tweet)))
tweets2
```

Had problems installing pybloom on mac. Found another filter called as dynamic pybloom in github with its functionality from here. https://github.com/srf5132/dynamic-pybloom/


```python
from dynamic_pybloom import ScalableBloomFilter
sbf = ScalableBloomFilter(mode=ScalableBloomFilter.SMALL_SET_GROWTH)
for tweet in tweets:
    sbf.add(tweet)
match = 0
nomatch = 0
for tweet in tweets2:
    if tweet in sbf:
        match += 1
    else:
        nomatch += 1
    
print(match)
print(nomatch)
```

Plotting a graph for matched vs not matched tweets between the two data sets

```python
%matplotlib inline
import matplotlib.pyplot as plt
x_vars = ['Match', 'Nomatch']
x_pos = list(range(len(x_vars)))
y_pos = [match,nomatch]
width = 0.6
fig, ax = plt.subplots()
plt.bar(x_pos, y_pos, width, alpha=1, color='g')
ax.set_ylabel('tweets', fontsize=15)
ax.set_xticks([p + 0.4 * width for p in x_pos])
ax.set_xticklabels(x_vars)
```
